```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113,50,45,20,110,120,90,80,65,55,72]
plt.hist(blood_sugar)
```

```python
blood_sugar=[113,50,45,20,110,120,90,80,65,55,72]
plt.hist(blood_sugar,bins=5)
```

```python
blood_sugar=[113,50,45,20,110,120,90,80,65,55,72]
plt.hist(blood_sugar,bins=5,rwidth=0.95)
```

```python
blood_sugar=[113,50,45,20,110,120,90,80,65,55,72]
plt.hist(blood_sugar,bins=[80,100,125,160],rwidth=0.95)
```

```python
blood_sugar=[113,50,45,20,110,120,90,80,65,55,72]
plt.hist(blood_sugar,bins=5,rwidth=0.95,color="g")
```

```python
blood_sugar=[113,50,45,20,110,120,90,80,65,55,72]
plt.hist(blood_sugar,bins=5,rwidth=0.95,color="g",histtype="step")
```

```python
blood_sugar_men=[113,50,45,20,110,120,90,80,65,55,72]
blood_sugar_women=[67,70,5,15,14,10,60,70,45,59,82]

plt.hist([blood_sugar_men,blood_sugar_women],bins=5,rwidth=0.95,color=["green","orange"])
```

```python
blood_sugar_men=[113,50,45,20,110,120,90,80,65,55,72]
blood_sugar_women=[67,70,5,15,14,10,60,70,45,59,82]

plt.hist([blood_sugar_men,blood_sugar_women],bins=5,rwidth=0.95,color=["green","orange"],label=["men","women"])
plt.legend()
```

```python
blood_sugar_men=[113,50,45,20,110,120,90,80,65,55,72]
blood_sugar_women=[67,70,5,15,14,10,60,70,45,59,82]

plt.xlabel("sugar range")
plt.ylabel("total no of patients")
plt.title("Blood sugar analysis")

plt.hist([blood_sugar_men,blood_sugar_women],bins=5,rwidth=0.95,color=["green","orange"],label=["men","women"])
plt.legend()
```

```python
blood_sugar_men=[113,50,45,20,110,120,90,80,65,55,72]
blood_sugar_women=[67,70,5,15,14,10,60,70,45,59,82]

plt.xlabel("sugar range")
plt.ylabel("total no of patients")
plt.title("Blood sugar analysis")

plt.hist([blood_sugar_men,blood_sugar_women],bins=5,rwidth=0.95,
         color=["green","orange"],label=["men","women"],orientation="horizontal"
        )
plt.legend()
```
