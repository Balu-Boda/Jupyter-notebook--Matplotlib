```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]
plt.pie(exp_val,labels=exp_labels)          
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]
plt.pie(exp_val,labels=exp_labels) 

plt.axis("equal")
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]
plt.pie(exp_val,labels=exp_labels) 

plt.axis("equal")

plt.show()
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]

plt.pie(exp_val,labels=exp_labels,radius=2) 

plt.axis("equal")

plt.show()
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]

plt.pie(exp_val,labels=exp_labels,radius=2,autopct="%0.2f%%") 

plt.axis("equal")

plt.show()
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]

plt.pie(exp_val,labels=exp_labels,radius=2,autopct="%0.0f%%") 

plt.axis("equal")

plt.show()
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]

plt.pie(exp_val,labels=exp_labels,radius=2,autopct="%0.0f%%",shadow=True) 

plt.axis("equal")

plt.show()
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]

plt.pie(exp_val,labels=exp_labels,radius=2,autopct="%0.0f%%",shadow=True,explode=[0,0,0,0,0]) 

plt.axis("equal")

plt.show()
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]

plt.pie(exp_val,labels=exp_labels,radius=2,autopct="%0.0f%%",shadow=True,explode=[0,0.2,0,0.1,0]) 

plt.axis("equal")

plt.show()
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]

plt.pie(exp_val,labels=exp_labels,radius=2,autopct="%0.0f%%",shadow=True,explode=[0,0.2,0,0.1,0],
       startangle=45) 

plt.axis("equal")

plt.show()
```

```python
exp_val=[100,150,180,120,190]
exp_labels=["A","B","C","D","F"]

plt.pie(exp_val,labels=exp_labels,radius=2,autopct="%0.0f%%",shadow=True,explode=[0,0.2,0,0.1,0],
       startangle=180) 

plt.axis("equal")

plt.show()
```
