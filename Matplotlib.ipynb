```python
import matplotlib.pyplot as mplt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[22,24,25,26,28,34,50]
```

```python
mplt.xlabel("days")
mplt.ylabel("temp")
mplt.title("Weather")
mplt.plot(x,y,color="green",linewidth=5,linestyle="dashdot")
```

```python
mplt.plot(x,y,"g+--")
```

```python
mplt.plot(x,y,"*r--")
```

```python
mplt.plot(x,y,"Dg-.")
```

```python
mplt.plot(x,y,color="r",marker="D",linestyle=" ")
```

```python
mplt.plot(x,y,color="green",marker="D",linestyle=" ",markersize=20)
```

```python
mplt.plot(x,y,color="green",marker="D",linestyle=" ",markersize=20,alpha=0.5)
```
