```python
import matplotlib.pyplot as plt
import numpy
%matplotlib inline
```

```python
company=["A","B","C","D"]
revenue=[100,90,200,50]
profit=[40,20,10,6]

ypos=numpy.arange(len(company))
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)

plt.ylabel("revenue(bln)")
plt.title("US companies")
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)

plt.ylabel("revenue(bln)")
plt.title("US companies")
plt.bar(ypos,revenue,label="revenue")
plt.legend()
```

```python
profit=[40,20,10,6]
plt.xticks(ypos,company)

plt.ylabel("revenue(bln)")
plt.title("US companies")
plt.bar(ypos,revenue,label="revenue")
plt.bar(ypos,profit,label="profit")
plt.legend()
```

```python
xpos=numpy.arange(len(company))
xpos
```

```python
profit=[40,20,10,6]
plt.xticks(ypos,company)

plt.ylabel("revenue(bln)")
plt.title("US companies")
plt.bar(xpos-0.2,revenue,label="revenue")
plt.bar(xpos+0.2,profit,label="profit")
plt.legend()
```

```python
profit=[40,20,10,6]
plt.xticks(ypos,company)

plt.ylabel("revenue(bln)")
plt.title("US companies")
plt.bar(xpos-0.2,revenue,width=0.4,label="revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

```python
plt.xticks(ypos,company)

plt.title("US companies")
plt.barh(xpos-0.2,revenue,label="revenue")
plt.barh(xpos+0.2,profit,label="profit")
plt.legend()
```
