```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[32,45,50,40,55,49,39]
min_t=[41,45,39,55,51,47,43]
avg_t=[33,46,53,57,52,48,56]
```

```python
plt.xlabel("days")
plt.ylabel("temp")
plt.title("Weather")
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
```
